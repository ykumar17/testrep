```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dpath =r'C:\Users\USER\Desktop\lpcmoirradiated\sorted'
fname_sub = r'NGOsubstrate-1021g-per_MH_E.dat'
fname_film = r'LPCMO-NGO-3-0891g-per_E.dat'
fpath_sub = os.path.join(dpath, fname_sub)
fpath_film = os.path.join(dpath, fname_film)
```

```python
names_sub = ["T(K)s ZFC", "M(emu/g)s ZFC 500Oe", "T(K)s FCCC", "M(emu/g)s FCC 500 Oe", "H(Oe)s 5K", "M(emu/g)s 5 K", "H(Oe)s 100 K", 
         "M(emu/g)s at 100 K","H(Oe)s 200 K", "M(emu/g)s at 200 K","H(Oe)s 300 K", "M(emu/g)s at 300 K"]
names_film = ["T(K)f ZFC", "M(emu/g)f ZFC 500Oe", "T(K)f FCCC", "M(emu/g)f FCC 500 Oe", "H(Oe)f 5K", "M(emu/g)f 5 K", "H(Oe)f 100 K", 
         "M(emu/g)f at 100 K","H(Oe)f 200 K", "M(emu/g)f at 200 K","H(Oe)f 300 K", "M(emu/g)f at 300 K"]

dfsub = pd.read_csv(fpath_sub,comment='!',  header = 0, names=names_sub )
dffilm = pd.read_csv(fpath_film, comment='!',  header = 0, names=names_film)
dffilm
```

```python
df_field = pd.concat([dffilm[[names_film[0], names_film[1]]],dfsub[[names_sub[0],names_sub[1]]],
                     dffilm[[names_film[2], names_film[3]]],dfsub[[names_sub[2],names_sub[3]]]] , axis =1)
#df_field.dropna(inplace = True)
df_field
```
